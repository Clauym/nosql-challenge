### 1. Which establishments have a hygiene score equal to 20?

```python
# Find the establishments with a hygiene score of 20
query = {'scores.Hygiene' : {'$eq': 20}}

# Use count_documents to display the number of documents in the result
count = establishments.count_documents(query)
print(f" \n Number of documents where Hygiene score equals 20: {count} \n")

# Display the first document in the results using pprint
results = establishments.find(query)
pprint(results[0])
```

```python
# Convert the result to a Pandas DataFrame
result_df = pd.DataFrame(list(results))
# Display the number of rows in the DataFrame
print(f"This dataframe contains {len(result_df.index)} rows")
# Display the first 10 rows of the DataFrame
result_df.head(10)
```

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.
query= {'LocalAuthorityName' : {'$regex': 'London'},'RatingValue': {'$gte': '4'}}

# Use count_documents to display the number of documents in the result
count=establishments.count_documents(query)
print(f" \n Number of documents where London is the Local authority: {count} \n")

# Display the first document in the results using pprint
results=establishments.find(query)
pprint(results[0])
```

```python
# Convert the result to a Pandas DataFrame
result_df = pd.DataFrame(list(results))
# Display the number of rows in the DataFrame
print(f"This dataframe contains {len(result_df.index)} rows")
# Display the first 10 rows of the DataFrame
result_df.head(10)
```

### 3. What are the top 5 establishments with a `RatingValue` rating value of '5', sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
# Sort by hygiene score

#Search for the coordinates of the restaurant
query={'BusinessName': 'Penang Flavours'}
fields = {'geocode'}

# Capture the results to variables
location = establishments.find_one(query, fields)["geocode"]
latitude = location["latitude"] 
longitude = location["longitude"]

print(f"Penang Flavours coordiantes: {latitude} lat , {longitude} long.")
```

```python
#Specify the maximum distances fom the location
degree_search = 0.01   
min_latitude=latitude - degree_search
max_latitude=latitude + degree_search
min_longitude=longitude - degree_search
max_longitude=longitude + degree_search

print(f"Location to search: ({round(min_latitude,5)}-{round(max_latitude,5)}) lat and ({round(min_longitude,5)}-{round(max_longitude,5)}) long.")
```

```python
# Query for coordinates and Ratings
coor_query = {'$and':[{'geocode.latitude' : {'$gte': min_latitude ,'$lte': max_latitude}},
                      {'geocode.longitude' : {'$gte': min_longitude ,'$lte': max_longitude}}],
                        'RatingValue':{'$eq' :'5'}}

#Selecting the fields to display
#fields={"BusinessName": 1, "RatingValue": 1, "scores.Hygiene":1, "geocode":1}

#Create a dictionay to sort
sort= [('scores.Hygiene', -1)]

#Set the number of restaurants wanted
limit=5

#Find the results
results = establishments.find(coor_query).sort(sort).limit(limit)
list_results=list(results)
# Print the results
pprint(list_results)
```

```python
# Convert result to Pandas DataFrame
pengang_df=pd.json_normalize(list_results)
pengang_df
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that: 
# 1. Matches establishments with a hygiene score of 0
match_query= {'$match': {'scores.Hygiene' : {'$eq': 0}}}

# 2. Groups the matches by Local Authority
group_query={'$group': {'_id': "$LocalAuthorityName", 'count': { '$sum': 1 }}}

# 3. Sorts the matches from highest to lowest
sort_values= {'$sort': { 'count': -1 }}

# Put the pipeline together
pipeline=[match_query, group_query, sort_values]

# Run the pipeline through the aggregate method and save the results to a variable
results = list(establishments.aggregate(pipeline))

# Print the number of documents in the result
documents=[ sub['count'] for sub in results ]

print('Number of restaurants with 0 Hygiene score: ', sum(documents))
print('Number of Local Authorities in the database: ', len(results), ' \n')

# Print the first 10 results
pprint(results[0:10])
```

```python
# Convert the result to a Pandas DataFrame
result_df = pd.DataFrame(results)

# Display the number of rows in the DataFrame
print(f'Dataframe number of rows: {len(result_df.index)}')

# Display the first 10 rows of the DataFrame
result_df.head(10)
```
